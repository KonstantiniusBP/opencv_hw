```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
img = cv2.imread('img/pexels.jpeg')

# ,   
if img is None:
    print("  ,    .")
else:
    #  BGR  RGB   
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    #  
    plt.imshow(img_rgb)
    plt.show()
```

```python
#     
blue, green, red = cv2.split(img)
```

```python
#      RGB
img1 = cv2.merge([red, green, blue])

#      RBG
img2 = cv2.merge([red, blue, green])

#      GRB
img3 = cv2.merge([green, red, blue])

#      BGR
img4 = img  #    BGR
```

```python
#  
out1 = np.hstack([img1, img2])
out2 = np.hstack([img3, img4])
out = np.vstack([out1, out2])

#  
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))  #    
plt.axis('off')
plt.show()
```

```python
#   
img_flipped_lr = np.fliplr(img)  #   
img_flipped_ud = np.flipud(img)  #   

#  
collage = np.hstack([img, img_flipped_lr])  #   
collage = np.vstack([collage, np.hstack([img_flipped_ud, np.fliplr(img_flipped_ud)])])  #   

#  
plt.imshow(cv2.cvtColor(collage, cv2.COLOR_BGR2RGB))  #    
plt.axis('off')
plt.show()
```
