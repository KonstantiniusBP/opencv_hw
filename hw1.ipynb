```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
```

```python
img = cv2.imread('img/pexels.jpeg')

# Проверим, что изображение загружено
if img is None:
    print("Изображение не найдено, проверьте путь к файлу.")
else:
    # Преобразование BGR в RGB для корректного отображения
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Отображение изображения
    plt.imshow(img_rgb)
    plt.show()
```

```python
# Разбиение изображения на цветовые каналы
blue, green, red = cv2.split(img)
```

```python
# Составление изображения в цветовой схеме RGB
img1 = cv2.merge([red, green, blue])

# Составление изображения в цветовой схеме RBG
img2 = cv2.merge([red, blue, green])

# Составление изображения в цветовой схеме GRB
img3 = cv2.merge([green, red, blue])

# Составление изображения в цветовой схеме BGR
img4 = img  # Оригинальное изображение в BGR
```

```python
# Создание коллажа
out1 = np.hstack([img1, img2])
out2 = np.hstack([img3, img4])
out = np.vstack([out1, out2])

# Отображение коллажа
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))  # Преобразование для правильного отображения
plt.axis('off')
plt.show()
```

```python
# Создание перевернутых изображений
img_flipped_lr = np.fliplr(img)  # Переворот по горизонтали
img_flipped_ud = np.flipud(img)  # Переворот по вертикали

# Создание коллажа
collage = np.hstack([img, img_flipped_lr])  # Первая строка коллажа
collage = np.vstack([collage, np.hstack([img_flipped_ud, np.fliplr(img_flipped_ud)])])  # Вторая строка коллажа

# Отображение коллажа
plt.imshow(cv2.cvtColor(collage, cv2.COLOR_BGR2RGB))  # Преобразование для правильного отображения
plt.axis('off')
plt.show()
```
