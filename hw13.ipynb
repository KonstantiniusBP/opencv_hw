## Loading training data and displaying a random sample

```python
import os
import cv2
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter

# Path to the Train.csv file
root = '.'  
data = pd.read_csv(os.path.join(root, 'Train.csv'))

# Checking the number of samples in the dataset
num_samples = len(data)
print(f"Number of samples in the dataset: {num_samples}")

# Displaying random images
plt.figure(figsize=(15, 10))
for ii in range(15):
    # Random index
    idx = np.random.randint(0, num_samples)
    # Loading the image
    img_path = os.path.join(root, data.iloc[idx]['Path'])
    img = cv2.imread(img_path)
    # Converting BGR (OpenCV) to RGB for displaying in matplotlib
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # Displaying the image
    plt.subplot(3, 5, ii + 1)
    plt.imshow(img)
    plt.title(f"Class: {data.iloc[idx]['ClassId']}")
    plt.axis('off')
plt.show()
```

## Inspecting the Dataset

```python
# Extracting class IDs
ids = data['ClassId']

# Counting the number of images in each class
hist = Counter(ids)

# Displaying the histogram
plt.figure(figsize=(12, 6))
plt.bar(hist.keys(), hist.values())
plt.grid(True)
plt.xlabel('Traffic Sign ID')
plt.ylabel('Counts')
plt.title('Per-Class Histogram')
plt.show()
```
